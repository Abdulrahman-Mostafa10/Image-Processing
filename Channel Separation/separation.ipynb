```python
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("../My Photo.png")

b, g, r = cv2.split(img)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(131)
plt.imshow(r, cmap="Reds", vmin=0, vmax=255)
plt.title("Red Channel")

plt.subplot(132)
plt.imshow(g, cmap="Greens", vmin=0, vmax=255)
plt.title("Green Channel")

plt.subplot(133)
plt.imshow(b, cmap="Blues", vmin=0, vmax=255)
plt.title("Blue Channel")
```

```python
merged_img = cv2.merge([r,g,b])
plt.figure(figsize=(12,6))

plt.subplot(121)
plt.imshow(merged_img)

plt.subplot(122)
gray_img = cv2.cvtColor(merged_img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap="gray")

plt.show()
```
